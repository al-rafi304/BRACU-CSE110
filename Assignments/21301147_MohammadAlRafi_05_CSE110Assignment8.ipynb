```python
file = open('TestFile.txt', mode='r')
lines = file.readlines()
new_lines = []
for line in lines:
    added_space = False
    modified_line = ''
    for char in line:
        if added_space and char == ' ':
            continue
        if char == ' ':
            added_space = True
        else:
            added_space = False
        modified_line += char
    new_lines.append(modified_line)
output_file = open('NewFile.txt', 'w')
output_file.writelines(new_lines)

output_file.close()
file.close()
```

Otherwise, print 'The sentence will work.'. 

**Note:** a,e,i,o,u are called vowels and all the other 21 english alphabets are consonants.

=============================

**Example 1:** \
**Input:**\
'his age is 10'\
**Output:**\
Number of vowels greater/equal to consonants. Please paraphrase.

=================

**Example 2:** \
**Input:**\
'he is 10 years old'\
**Output:**\
The sentence will work.

```python
string = input("Enter a sentence: ").lower()
vowels = ['a','e','i','o','u']
vowel_count = 0
cons_count = 0
try:
    for char in string:
        if ord(char) >= 97 and ord(char) <= 122:
            if char in vowels:
                vowel_count += 1
            else:
                cons_count += 1
    if vowel_count == cons_count:
        raise RuntimeError
except RuntimeError:
    print("Number of volwels greater/equal to consonants. Please paraphrase.")
else:
    print("This sentence will work.")
```
