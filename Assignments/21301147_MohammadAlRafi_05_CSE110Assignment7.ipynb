### Task 1

Suppose you have a list named my_list as given below. <br/>
my_list = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29] <br/>
Now use bubble sort to sort my_list into ascending order.

**Output**<br/>
[1, 2, 2, 3, 5, 6, 10, 11, 12, 14, 15, 17, 18, 20, 29]

```python
def bubble_sort(data):
  for count1 in range(len(data) - 1):
    for count2 in range(len(data) - 1 - count1):
      if data[count2] > data[count2+1]:
        temp = data[count2]
        data[count2] = data[count2 + 1] 
        data[count2 + 1] = temp

my_list = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29]
bubble_sort(my_list)
print(my_list)
```

### Task 2

Suppose you have a list named my_list as given below. <br/>
my_list = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29] <br/>
Now use selection sort to sort my_list into ascending order.

**Output**<br/>
[1, 2, 2, 3, 5, 6, 10, 11, 12, 14, 15, 17, 18, 20, 29]



```python
def selection_sort(data):
    result = []
    result.extend(data)

    for index in range(len(result)):
        min = index

        for count in range(index+1, len(result)):
            if result[min] > result[count]:
                min = count
        
        if result[min] < result[index]:
            temp = result[min]
            result[min] = result[index]
            result[index] = temp
    
    return result

my_list = [10,1,20,3,6,2,5,11,15,2,12,14,17,18,29]
sorted_list = selection_sort(my_list)
print(sorted_list)
```

### Task  3

Suppose you have a list named my_list as given below <br/>

Now use any of the two sorting techniques you have used above to sort the list in descending order.

**Output**<br/>
[29, 20, 18, 17, 15, 14, 12, 11, 10, 6, 5, 3, 2, 2, 1]


```python
def descending_order_sort(data):
    result = []
    result.extend(data)

    for index in range(len(result)):
        max = index

        for count in range(index+1, len(result)):
            if result[max] < result[count]:
                max = count
        
        if result[max] > result[index]:
            temp = result[max]
            result[max] = result[index]
            result[index] = temp
    
    return result

my_list=[10,1,20,3,6,2,5,11,15,2,12,14,17,18,29]

result = descending_order_sort(my_list)
print(result)
```

### Task 4

Suppose you have a sitting arrangement of the students who will give the final exam in a list. The list contains the last two digits of their student ID as given below.<br/>

sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80] <br/>

Now you want to organize the sitting arrangement of the students in your own way. You decide to sort all the students in the even indices of the list in ascending order and all the students in the odd indices of the list in descending order.  <br/>

So, write a python program that organizes the list for you in this way.<br/>


**Sample Output**<br/>
[10, 100, 11, 80, 12, 70, 20, 56, 22, 30, 58, 16, 70, 15]

```python
def custom_sort(data):
  #Creates two list containing even and odd indexed values
  even_list = []
  odd_list = []
  for index in range(len(data)):
    if index % 2 == 0:
      even_list.append(data[index])
    else:
      odd_list.append(data[index])
  
  #Sorts both lists
  sorted_even = sorted(even_list)
  sorted_odd = sorted(odd_list, reverse=True)

  #Creates a new organized list from sorted_even and sorted_odd
  result = []
  count = 0
  even_count = 0
  odd_count = 0
  while count <= len(sorted_even) + len(sorted_odd) - 1:
    if count % 2 == 0:
      result.append(sorted_even[even_count])
      even_count += 1
    else:
      result.append(sorted_odd[odd_count])
      odd_count += 1
    count += 1
  return result

sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]
result = custom_sort(sitting_list)
print(result)
```

### Task 6

Suppose you have a list named my_list as given below. Your task is to sort the list in ascending order and print the count of numbers that have changed their positions in the process of sorting. <br/>

my_list = [4, 2, 3, 1, 6, 5] <br/>

Sorted list would be [1, 2, 3, 4, 5, 6] where 4 numbers (4, 1, 6 and 5) have changed their positions. Therefore our sample output here would be 4. Please check if your code is working correctly by changing the above list and using the knowledge given here to verify whether your code gives correct output for all different lists.


**Sample Output for the above list**<br/>
4

```python
def count_swaps(data):
  result = []
  result.extend(data)
  
  #Sorting list
  for index in range(len(result)):
    min = index

    for count in range(index+1, len(result)):
      if result[min] > result[count]:
        min = count
        
    if result[min] < result[index]:
      temp = result[min]
      result[min] = result[index]
      result[index] = temp
  
  #Checking swap count
  swap_count = 0
  for index in range(len(data)):
    if data[index] != result[index]:
      swap_count += 1
  return swap_count

my_list = [4, 2, 3, 1, 6, 5]
result = count_swaps(my_list)
print(result)
```

### Task 7

Write a python program that takes two lists from the user, merges the two lists, sorts the resulting list, and then finds the median of the elements in the two lists.

=====================================================


**Sample Input 1**<br/>
list_one = [1, 2, 1, 4] <br/>
list_two = [5, 4, 1]


**Sample Output 1**<br/>
Sorted list = [1, 1, 1, 2, 4, 4, 5] <br/>
Median = 2

=====================================================


**Sample Input 2**<br/>
list_one = [1, 7, 9, 10] <br/>
list_two = [2, 7, 6, 5]


**Sample Output 2**<br/>
Sorted list = [1, 2, 5, 6, 7, 7, 9, 10] <br/>
Median = 6.5



```python
def find_median(data):
  median_index = 0
  if len(data) % 2 != 0:
    median_index = data[len(data) // 2]
  else:
    median_index = (data[(len(data) // 2)] + data[((len(data)//2) - 1)]) / 2
    
  return median_index

list_one = [1, 7, 9, 10]
list_two = [2, 7, 6, 5]

new_list = list_one + list_two
new_list.sort()
print("Sorted list =", new_list)
print("Median =", find_median(new_list))
```
