### Task 9

Write a Python program that reads 10 numbers into a list and prints the numbers of that specified list after removing **even numbers** from it.

=========================================================================

**Sample Input:**

7,12,4,55,96,2,11,61,33,42

[7, 55, 11, 61, 33]

=========================================================================





```python
input_list = []
#Adding inputs to a list
for count in range(10):
  number = int(input("Enter a number: "))
  input_list.append(number)

index = 0
while index < len(input_list):
  current = input_list[index] 
  if current % 2 == 0:
    input_list.remove(current)
    index = 0
  else:
    index += 1

print(input_list)
```

### Task 15


Write a Python program that reads 5 numbers into a list and prints the second largest number and its location or index position on the list. <font color='red'>[You are not allowed to use the max(), sort(), sorted() function here]</font>

===================================================================

**Sample Input:**\
 7, 13, 2, 10, 6

**Sample Output:**\
My list: [7, 13, 2, 10, 6]\
Second largest number in the list is 10 which was found at index 3.

===================================================================

```python
input_list = list(map(int, input("Enter your first number list: ").split(",")))
sorted_list = list(input_list)

#Storting list in ascending order
for item in sorted_list:
  for index in range(1, len(sorted_list)):
    current = sorted_list[index]
    prev = sorted_list[index - 1]

    if current < prev:
      #Swaps current and previous items
      sorted_list[index] = prev
      sorted_list[index - 1] = current

max = sorted_list[-2]   #Because we need second last item to get second largest value in a ascending order list
print("My list:", input_list)
print("Second largest number in the list is " + str(max) + " which was found at index " + str(input_list.index(max)))
```
