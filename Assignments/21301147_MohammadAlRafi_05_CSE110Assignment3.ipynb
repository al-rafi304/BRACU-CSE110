=====================================================================

**Example 1:**\

```python
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")
new_string = ""
total_len = len(string1) + len(string2)

for count in range(total_len):

  #Adds the remaining characters from string2
  if count > len(string1) - 1:
    new_string += string2[count:]
    break
  #Adds the remaining characters from string1
  elif count > len(string2) - 1:
    new_string += string1[count:]
    break
  #Adds charecters from string1 and string2 one after another
  else:
    new_string += string1[count] + string2[count]

print(new_string)
```

### Task 6
Given a string, create a new string with all the consecutive duplicates **removed**. 

=====================================================================

**Hint:** You may make a new string to store the result. You can check whether the current character and the next character are the same, then add that character to the new string.

=====================================================================


**Sample Input**:\
AAABBBBCDDBBECE

**Sample Output:**\
ABCDBECE

=====================================================================


```python
string = input("Enter a string: ")
new_string = ''
for char_count in range(len(string)):

  #Adds final charecter without checking because previous concurrent characters will be skipped
  if char_count == len(string) - 1:
    new_string += string[char_count]
    break

  #Skips duplicates if upcoming character matches with current one
  if string[char_count] == string[char_count + 1]:
    continue
  #Adds characters to new string
  else:
    new_string += string[char_count]

print(new_string)
```

```python
string = input("Enter a String: ")
make_upper = True       #Because string must start will uppercase
new_string = ''

for char in string:

  if make_upper:
    new_string += char.upper()
    #Skips changing state if there is a space between characters
    if char == " ":
      continue
    make_upper = False
  else:
    new_string += char.lower()
    #Skips changing state if there is a space between characters
    if char == " ":
      continue
    make_upper = True
  
print(new_string)
```
