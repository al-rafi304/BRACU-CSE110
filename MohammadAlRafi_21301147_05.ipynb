# Question 2

```python
string = input("Enter a string: ")
number = int(input("Enter a number:"))

for index in range(len(string)):
  if number % 2 == 0 and index % 2 == 0:
    print(chr(ord(string[index]) + 1 - 32), end='')
  elif number % 2 != 0 and index % 2 != 0:
    print(chr(ord(string[index]) + 1 - 32), end='')

```

# Question 1


```python
number = int(input("Enter a number: "))

for count in range(1, number+1):
    print("=== Iteration " + str(count) + " ===")
    if count % 2 != 0:
        line = ''
        for count in range(1, count+1):   #Storing numbers sequentially 
            line += str(count)
        for row in range(1, count+1):     #printing sequenced numbers line by line
            print(line)
    else:
        for row in range(1, count+1):
            print("X" * row)
    
```
