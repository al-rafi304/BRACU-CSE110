# Question 1:

```python
cast = {'Friends': [('Joey', 'Matt'), ('Chandler', 'Matthew')], 
        'BBT': [('Penny', 'Kaley'), ('Sheldon', 'Jim')], 
        'Breaking Bad': [('Jesse', 'Aaron'), ('Walter', 'Bryan')]}
output = ''
series = ''
character = ''
name = input("Enter a name of charecter: ")
for key in cast.keys():
    for chars in cast[key]:
        if name in chars:
            series = key
            character = chars[0]
            break

output = name + ' played the character ' + character + ' in ' + series
print(output)
```

# Question 2:

```python
given_data = [["Alice", ["A"]], ["Sam", [ "A", "E", "B"]]]

marksheet = {'A': 5, 'B': 4, 'C': 3, 'D': 2, 'E': 0}

for student in given_data:
    if len(student[-1]) < 2:
        print(student[0] + ': You must have at least 2 subjects in A level.')
        continue
    total = 0
    average = 0
    marks = []
    for mark in student[-1]:
        marks.append(marksheet[mark])
    sorted_marks = sorted(marks)
    if sorted_marks[-1] == 0 or sorted_marks[-2] == 0:
        print('Sorry ' + student[0] + ' the grades allowed in your top 2 grades are only A/B/C/D')
        continue
    total = sorted_marks[-1] + sorted_marks[-2]
    average = total / 2
    print('Based on the A level result ' + student[0] + "'s minimum GPA is " + str(average))
```

# Question 3:

```python
def replacer(string, replacer):
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    modified_string = ''
    for char in string:
        if char.lower() in vowels and replacer == 'vowels':
            modified_string += 'V'
            continue
        elif char.lower() not in vowels and ord(char.lower()) in range(97, 123) and replacer == 'consonants':
            modified_string += 'C'
            continue
        modified_string += char
    return modified_string

print(replacer("It is MAT110 final", "vowels"))
```
