```python
def check_magic(num):
  even_sum = 0
  odd_sum = 0
  for index in range(len(str(num))):
    if index % 2 == 0:
      even_sum += int(num[index])
    else:
      odd_sum += int(num[index])
  
  return True if even_sum == odd_sum else False

numbers = "442310, 441310, 42251, 42241, 451, 452" #input("Enter numbers: ")
magic = []
non_magic = []
for number in numbers.split(','):
  if check_magic(number.strip()):
    magic.append(int(number))
  else:
    non_magic.append(int(number))

result = (tuple(magic), tuple(non_magic))
print(result)
```

```python
def get_data(string):
    index = 1
    result = []
    while index < len(string):
        prev = string[index - 1]
        
        if prev == '(':
            sub_items = ''
            while string[index] != ')':
                sub_items += string[index]
                index += 1
            result.append(sub_items)
        index += 1
    return result

def get_sum(string):
    sum = 0
    for char in string.split(","):
        sum += int(char.strip())
    return sum
    
given = "(33, 22, 11), (30, 45, 56, 45,20), (81, 90, 39, 45), (1, 2, 3, 4,5,6)" #input("Enter a tuple of tuples: ").split(",")
cleared_data = get_data(given)
average_list = []
max_list = []
max = get_sum(cleared_data[0])
for items in cleared_data:
    avrg = 0
    sum = 0
    count = 0
    for item in items.split(','):
        sum += int(item.strip())
        count += 1
    avrg = sum / count 
    average_list.append(avrg)
    
    if sum > max:
        max_list = [int(num.strip()) for num in items.split(',')]

print(average_list , tuple(max_list))

```

```python
def check_prime(num):
    for count in range(2, num):
        if num % count == 0:
            return False
    return True

def return_prime_list(list1):
    prime_list = []
    for item in list1:
        if check_prime(item):
            prime_list.append(item)
    return prime_list

given = {"N":(4,9,3),"k":[95,37,197],"F":(32,5,97),"s":[31,94,55]}
dict_primes = {}
for key, value in given.items():
    if key.islower():
        dict_primes.update({key: return_prime_list(value)})
    else:
        dict_primes.update({key: tuple(return_prime_list(value))})

print(return_prime_list([5,2,55,17]))
print(dict_primes)
```
