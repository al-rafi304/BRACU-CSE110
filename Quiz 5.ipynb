```python
def string_modify(string):
    vowels = ['a,','e','i','o','u']
    last = string[-1]

    if len(string) <= 1:
        return string

    if last in vowels:
        return 'yay' + string
    else:
        temp = ''
        for index in range(len(string) - 1):
            temp += string[index]
        return last + temp + 'ay'

result = string_modify('a')
print(result)
```
