### Task 1
Write a python program that will take a number as an input and check whether the number is Awesome or not. If the number is Awesome, it prints True. Otherwise False.

**Awesome number**: a number where every digit is less than its immediate left digit is called an Awesome number. A single-digit number cannot be an awesome number(e.g. 5421 is an Awesome number.)

==========================================================

**Sample Input1:**\
976321\
**Sample Output1:**\
True

==========================================================

**Sample Input2:**\
9766321\
**Sample Output2:**\
False

==========================================================

**Sample Input3:**\
9\
**Sample Output3:**\
False

==========================================================

**Sample Input4:**\
78\
**Sample Output4:**\
False

==========================================================

**Sample Input5:**\
87\
**Sample Output4:**\
True


```python
numbers = input("Enter number: ")

if len(numbers) > 1:
    is_awesome = True
    for index in range(len(numbers) - 1):
        curr = int(numbers[index])
        next = int(numbers[index + 1])
    
        if curr < next or curr == next:
            is_awesome = False
            break
else:
    is_awesome = False

print(is_awesome)
```

### Task 2

Write a python program that will take a number a list of numbers(integer) as an input from the user and print whether the number is Awesome or not. 

==========================================================

**Sample Input1:**\
[976321, 321, 9763221, 9742, 876, 3211]\
**Sample Output1:**\
976321 is an awesome number.\
321 is an awesome number.\
9763221 is a not-so-awesome number.\
9742 is an awesome number.\
876 is an awesome number.\
3211 is a not-so-awesome number.

==========================================================

**Sample Input2:**\
[97821, 97210, 979210]\
**Sample Output2:**\
97821 is a not-so-awesome number.\
97210 is an awesome number.\
979210 is a not-so-awesome number.


```python
number_list = input("Enter a list of number: ").replace(" ", "").split(",")

for item in number_list:
    numbers = item

    if len(numbers) > 1:
        is_awesome = True
        for index in range(len(numbers) - 1):
            curr = int(numbers[index])
            next = int(numbers[index + 1])
    
            if curr < next or curr == next:
                is_awesome = False
                break
    else:
        is_awesome = False
    
    print(numbers + " is an awesome number" if is_awesome else numbers + " is a not-so-awesome number")
```

### Task 4

Write a python program that will take a number a list of numbers(integer) as an input from the user and does the following.
* counts the number of valid, invalid, and duplicate numbers and prints these to the user.
* stores the valid and invalid numbers in a list format and prints these to the user. You need to make sure that a number is stored only Once. **Valid numbers:** A number with a minimum length of 4 and a maximum length of 6 is considered to be valid for this assignment.
* Calculate and print the percentage of valid numbers, invalid numbers, and duplicate numbers to the user. 


**Hints:** Use membership operator in and not in for checking the duplicates. For calculating the percentage, count the numbers of a particular number type.

**Sample Input1:**\
[97821, 1, 97210, 963, 979210, 979210]\

Valid numbers in list: [976321, 9742, 3211]\
Invalid number count: 3\
Invalid numbers in list: [321, 9763221, 876]\

================================================\
Valid number percentage(%): 37.5\
Invalid number percentage(%): 37.5\
Duplicate number percentage(%): 25.0\
================================================\


```python
number_list = input("Enter a list of number: ").replace(" ", "").split(",")
valid_count = 0
invalid_count = 0
duplicate_count = 0
uniq_list = []
valid_list = []
invalid_list = []

#Finding and clearing duplicates
for item in number_list:
    if item not in uniq_list:
        uniq_list.append(item)
    else:
        duplicate_count += 1

#Finding valid, invalid numbers
for item in uniq_list:
    if len(item) >= 4 and len(item) <= 6:
        valid_count += 1
        valid_list.append(item)
    else:
        invalid_count += 1
        invalid_list.append(item)

print("================================================")
print("Total numbers present in the given string:", len(number_list))
print("Valid number count:", valid_count)
print("Valid numbers in list: ", valid_list)
print("Invlid number count:", invalid_count)
print("Invalid numbers in list: ", invalid_list)
print("Duplicate number count:", duplicate_count)
print("================================================")
print("Valid number percentage(%):", (len(valid_list)/len(number_list)) * 100)
print("Invalid number percentage(%):", (len(invalid_list)/len(number_list)) * 100)
print("Duplicate number percentage(%):", (duplicate_count/len(number_list)) * 100)
print("================================================")


```

Finally, print the nested list to the user.

================================================


```python
number_list = input("Enter a list of number: ").replace(" ", "").split(",")
new_list = []
for item in number_list:
    individual_data = []
    sum = 0
    product = 1
    #Check if valid or not
    if len(item) >= 4 and len(item) <= 6:
        for num in item:
            sum += int(num)
            product *= int(num)

        individual_data.append(int(item))
        individual_data.append(sum)
        individual_data.append(product)

        new_list.append(individual_data)

print("Data [number, sum of digits, product of digits]:",new_list)
```
