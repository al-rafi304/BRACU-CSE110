# 11

```python
strings = input("Enter a string").lower().split(" ")        #Note: We need to lowercase it because if one charecter is uppercase, code won't work
new = ""
for item in strings:
    new += chr(ord(item[0]) - 32) + item[1:] + " "

print(new)
```

# 11 Alternative(without using list)

```python
strings = input("Enter a string").lower()
new = ""
capital = True      #First letter need to be capitalized always
for char in strings:
    if char == " ":     
        capital = True      #After a space we need to capitalize the next charecter
        new += char
        continue
    if capital:
        char = chr(ord(char) - 32)      #every charecter's Capitalized version is 32 behind itself
        new += char
        capital = False         #We don't need to capitalize next charecter
        continue
    new += char

print(new)
```

```python
list01 = list(input("Enter a list: ").split(","))
new = []

for item in list01:
    new.append(item[::-1])          #Steps is -1 so, it will go like -1,-2,-3.... hence, reversing the list

print(new)
```

# 15

```python
list01 = list(map(int, input("Enter a list: ").split(",")))

even = 'not found'
odd = 'not found'

for item in list01:
    if item % 2 == 0:
        even = item
        break           #Breaking coz we only need the first occurrence

for item in list01:
    if item % 2 != 0:
        odd = item 
        break           #Breaking coz we only need the first occurrence

print("First even:" + str(even) + " and First odd:" + str(odd))
```

```python
a = list(input("Enter list: ").split(","))
n = int(input("Enter value of N:"))
new = []
for i in range(1,n+1):
    for item in a:
        new.append(item + str(i))

print(new)
```
